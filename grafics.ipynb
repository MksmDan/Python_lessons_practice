```python
import matplotlib.pyplot as plt
import math
from matplotlib.patches import Circle, Wedge, Polygon, Ellipse, Arc, Path, PathPatch
```

```python
def f_x(x):
    y = x ** 3 - 6 * x ** 2 +  x + 5
    return y
```

```python
a = -2
b = 6
n = 100

h = (b - a) / (n - 1)
```

```python
x_list = [a + h * i for i in range(n)]

f_list = [f_x(x) for x in x_list]
```

```python
line = plt.plot(x_list, f_list)

plt.setp(line, color="blue", linewidth=2)
plt.gca().spines["left"].set_position("zero")
plt.gca().spines["bottom"].set_position("zero")

plt.show()
```

```python
def compute_population(t):
    C = 172 
    T1 = 2000
    tay = 45
    N = C / tay * (math.pi / 2 - math.atan((T1 - t) / tay))                
    return N
```

```python
a = int(input())
b = int(input())

if compute_population(b) < 10.5 or compute_population(a) > 10.5:
    print('В данном временном промежутке численность населения менее 10.5 млрд людей')
    
else:
    x_list = [i for i in range(a, b)]
    y_list = [compute_population(i) for i in x_list]
    
    line = plt.plot(x_list, y_list)
    
    plt.gca().spines["left"].set_position("center")
    plt.gca().spines["bottom"].set_position("center")
    plt.gca().spines["top"].set_visible(False)
    plt.gca().spines["right"].set_visible(False)
    
   
```

```python
def f_x(x):
    x = x * math.pi / 180
    f = math.exp(math.cos(x)) +math.log(math.sin(0.8 * x) ** 2 + 1) * math.cos(x)
    
    return f
```

```python
def y_x(x):
    x = x * math.pi / 180
    y = -math.log((math.cos(x) + math.sin(x)) ** 2 + 1.7) + 2
    
    return y
```

```python
a = -240 
b = 360
n = 1000
h = (b - a) / (n - 1)

x_list = [a + h * i for i in range (n) ]
f_list = [f_x(x) for x in x_list]
y_list = [y_x(x) for x in x_list]

line_f = plt.plot(x_list, f_list, label='f(x)')
line_y = plt.plot(x_list, y_list, label='y(x)')

plt.setp(line_f, color="blue", linewidth=2)
plt.setp(line_y, color="red", linewidth=2)

plt.gca().spines["left"].set_position("center")
plt.gca().spines["bottom"].set_position("center")
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

plt.legend()
plt.show
```
